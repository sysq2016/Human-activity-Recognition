```python
#set the path of os
path = 'C:\\Users\\Administrator\\Desktop\\141b-hw3-dandipeng\\sta141b-hw3-dandipeng\\data\\fruit'
os.chdir(path)
files_frt = os.listdir(path)
files_frt
```

```python
# read all excel files into a data frame - 'fruit'

fruit = pd.DataFrame()
for i in files_frt:
    data = pd.read_excel(i,header = None)
    fruit = fruit.append(data)
```

```python
# extra rows of fruit names 
fruit_name = fruit.loc[fruit.index.isin([0])]

#create a column of type
tp_frt = pd.DataFrame({'type': ['fruit'] * len(fruit_name)})

#extra rows of 'Fresh' and 'Prices'
fruit_price = fruit.loc[fruit.index.isin([3])]

#drop columns that are all 'NaN's
fruit_name = fruit_name.dropna(axis=1, how='all')
fruit_price = fruit_price.dropna(axis=1, how='all')
```

```python
# add names for fruit_name
colnames1 = ['food']
fruit_name = fruit_name.rename(columns={i:cn for i,cn in enumerate(colnames1)})

# reset index of 'fruit_name' to be convenient to concarate data frames later 
fruit_name = fruit_name.reset_index(drop=True)

# only reserve the fruit names for the fruit_name data frame
for i in range(len(fruit_name)):
    for j in range(len(fruit_name.iloc[i,0])):
        if fruit_name.iloc[i,0][j] == '—':
            fruit_name.at[i, 'food'] = fruit_name.iloc[i,0][:j]
            break
```

```python
# reset index of 'fruit_price' to be convenient to concarate data frames later 
fruit_price = fruit_price.reset_index(drop=True)
fruit_fnl = pd.concat([tp_frt, fruit_name, fruit_price], ignore_index=True, axis=1)

# add names for fruit_fnl
colnames2 = ['type','food', 'form', 'price_per_lb', '_', 'yield', 'lb_per_cup', '_', 'price_per_cup']
fruit_fnl= fruit_fnl.rename(columns={i:cn for i,cn in enumerate(colnames2)})

# drop columns with measurements in 'fruit_fnl'
fruit_fnl = fruit_fnl.drop(['_'],axis =1)
fruit_fnl = fruit_fnl[fruit_fnl['form'].str.contains("Fresh") == True]
fruit_fnl = fruit_fnl.reset_index(drop=True)
fruit_fnl
```
